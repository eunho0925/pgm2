```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("penguins")
df
```

```python
plt.scatter(df['bill_length_mm'], df['bill_depth_mm'])
plt.xlabel('bill length')
plt.ylabel('bill depth')
```

```python
df['species']
```

```python
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])
plt.xlabel('flipper_length')
plt.ylabel('body_mass')
```

```python
# 체중 데이터 스케일링 (예시)
min_weight = 10
max_weight = 100
scaled_weights = (df['body_mass_g'] - df['body_mass_g'].min()) / (df['body_mass_g'].max() - df['body_mass_g'].min())
sizes = min_weight + scaled_weights * (max_weight - min_weight)

plt.scatter(df['flipper_length_mm'], df['body_mass_g'], s=sizes)
plt.xlabel('flipper_length')
plt.ylabel('body_mass')
```

```python
# 체중 데이터 스케일링 (예시)
min_weight = 10
max_weight = 100
scaled_weights = (df['body_mass_g'] - df['body_mass_g'].min()) / (df['body_mass_g'].max() - df['body_mass_g'].min())
sizes = min_weight + scaled_weights * (max_weight - min_weight)

plt.scatter(df['flipper_length_mm'], df['body_mass_g'], s=sizes, c=df['flipper_length_mm']/10, cmap='plasma', alpha=0.7)
plt.xlabel('flipper_length')
plt.ylabel('body_mass')
```

```python
# 체중 데이터 스케일링 (예시)
min_weight = 10
max_weight = 100
scaled_weights = (df['body_mass_g'] - df['body_mass_g'].min()) / (df['body_mass_g'].max() - df['body_mass_g'].min())
sizes = min_weight + scaled_weights * (max_weight - min_weight)

plt.figure(figsize=(7, 7))

plt.scatter(df['flipper_length_mm'], df['body_mass_g'], s=sizes, c=df['flipper_length_mm'], cmap='viridis', alpha=0.7)
plt.xlabel('flipper_length')
plt.ylabel('body_mass')
plt.colorbar(label='flipper_length', shrink=0.5, orientation='horizontal')
plt.title('펭귄의 지느러미와 체중의 상관관계')
plt.show()
```
